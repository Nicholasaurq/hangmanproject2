```python
def blankReplace(word):
    #create a string with just underscores that equal the length of the word being guessed
    wordBlank = "_"*len(word)
    #string converted to list
    wordBlank = list(wordBlank)
    return wordBlank
def printWord(wordProg):
    #rather than printing the list, which has commas and brackets,
    #thought it would be cleaner just to have spaces in between each letter
    stringWord = ""
    for c in wordProg:
        stringWord = stringWord + " " + c
    print(stringWord)
def decrypt(ciphertext):
    plaintext = ""
    i = 0
    #encrypt took the value of index to determine if the ascii value gets added or subtracted by 1
    #decrypt just reverses this, determines if i is even or odd, then converts character to ascii value, add/sub 1, then add it to plain text
    for c in ciphertext:
        if i % 2 == 0:
            x = ord(c) - 1
            plaintext += chr(x)
        else:
            x = ord(c) + 1
            plaintext += chr(x)
        i += 1
    return plaintext
```

```python
import random

#variables created
print("Welcome to the hangman game!")
errors = 0

#words list has a list of words to choose from. using a random number, word gets assigned a word
words = ["WHDSPQZ", "XHO", "TTDBFRT", "QQJYF", "ENQD", "DNPK", "CNTR", "EHWHOD", "TSVMOHOF", "XNOCFQGTM"]
word = words[random.randint(0, 9)]

#the word is converted to a list for better iteration and evaluation, word string still kept
wordChar = list(word)

#wordProg keeps track of the users progress at guessing the word, it shows the letters you have guessed while keeping the rest blank
#its also used to make sure the user isn't guessing with the same letter
wordProg = blankReplace(word)
printWord(wordProg)

#while loop continues unless 10 errors are reached or there are no more empty spaces left in wordProg
while errors != 10 and '_' in wordProg:
    #variables created, index for lists, total to count how many spaces are filled with each guess, and found to determine if user guessed a correct letter
    index = 0
    total = 0
    found = 0
    user = input("Enter your guess: ")
    #each item in word char is evaulated, first determine if the letter has already been guessed
    for c in wordChar:
        if user == wordProg[index]:
            print("You have already guessed that letter!")
        #next guess if the letter is apart of the word
        elif user == c:
            wordProg[index] = c
            found = 1
            total += 1
        #if neither are true, nothing happens, meanign found is never 1, meaning user gets an error
        else:
            print()
        index += 1
    if found == 1:
         print("you found " + str(total) + " letter(s)!")
    else:
        print("Incorrect")
        errors+=1
    print("Errors: " + str(errors))
    printWord(wordProg)

#when the while loop is exited, determines if the user won or lost, if one decrypt method starts
if found == 1:
    print("Congrats! You solved it!")
    print("The Decrypted word is: " + decrypt(word))
else:
    print("You lose!")
    
```
